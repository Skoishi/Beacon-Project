```python
# this file takes in the analyzer data, and made it into plot
# it plot out the V_rms vs time in the day
# it will also fit the data into a sine wave, and show the galaxy altitude in given date
# idealy if most of the anthropengic noise getting removed, the final fit should be in phase respect to galaxy altitude 


import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from Astropy import get_galaxy_altitudes
import os
from datetime import datetime, timedelta
from astropy.time import Time 
import matplotlib.dates as mdates

# Example filename
path="sample_data/"
filenames = ["STD_plot_2023-11-30 00:00:34_channel_4_random_median_filtered.txt",
           "STD_plot_2023-11-30 23:58:42_channel_4_random_median_filtered.txt",
           "STD_plot_2023-12-02 00:01:21_channel_4_random_median_filtered.txt",
           "STD_plot_2023-12-02 23:59:18_channel_4_random_median_filtered.txt",
           "STD_plot_2023-12-03 23:58:26_channel_4_random_median_filtered.txt",
           "STD_plot_2023-12-04 23:58:10_channel_4_random_median_filtered.txt",
           "STD_plot_2023-12-06 00:00:12_channel_4_random_median_filtered.txt",
           "STD_plot_2023-12-07 00:00:50_channel_4_random_median_filtered.txt"]
#filenames = ["STD_plot_2024-03-30 00:01:02_channel_10_random_median_with20.txt"]

completefile=[os.path.join(path, filename) for filename in filenames]

class Event_day:
    def __init__(self, readout_time_array, V_rms_array, error_bar_array):
        self.readout_time = readout_time_array
        self.V_rms = V_rms_array
        self.error_bar = error_bar_array
        #self.V_rms_20 = V_rms_20_array


event_day_arr=[]

for filename in completefile:
    # Initialize arrays
    readout_time_array = []
    V_rms_array = []
    error_bar_array = []
    #V_rms_20_array=[]

    
    # Open the file for reading
    with open(filename, 'r') as file:
        lines = file.readlines()
        current_array = None  # This will keep track of which array we are currently reading into
        
        for line in lines:
            # Remove any leading/trailing whitespace
            line = line.strip()
    
            # Check if the line indicates the start of a new array
            if line.startswith("Readout Time:"):
                current_array = "readout_time"
            elif line.startswith("V_rms Median:"):
                current_array = "V_rms"
            elif line.startswith("Error Bars (Lower, Upper):"):
                current_array = "error_bars"
            elif line.startswith("V_rms at lowest 20%:"):
                current_array = "V_rms_20"
            elif line and current_array:  # Process data lines only if current_array is set
                try:
                    if current_array == "readout_time":
                        readout_time_array.extend([float(x.strip()) for x in line.split(",") if x.strip()])
                    elif current_array == "V_rms":
                        V_rms_array.extend([float(x.strip()) for x in line.split(",") if x.strip()])
                    #elif current_array == "V_rms_20":
                        #V_rms_20_array.extend([float(x.strip()) for x in line.split(",") if x.strip()])
                    elif current_array == "error_bars":
                        error_pairs = line.split("),")
                        for pair in error_pairs:
                            pair = pair.replace("(", "").replace(")", "").strip()
                            lower, upper = map(float, pair.split(","))
                            error_bar_array.append((lower, upper))
                except ValueError:
                    break
                    #print(f"Skipping line due to conversion error: {line}")

    # Convert lists to numpy arrays
    t = np.array(readout_time_array)
    V_rms = np.array(V_rms_array)
    #V_rms_20 = np.array(V_rms_20_array)
    error_bars = np.array(error_bar_array)

    #event_day_arr.append(Event_day(t,V_rms,error_bars,V_rms_20))
    event_day_arr.append(Event_day(t,V_rms,error_bars))
# Print the first few entries to verify
print("V_rms Array:", event_day_arr[0].readout_time[:10])
#print("Readout Time Array:", t[:10])
#print("Error bar Array:", error_bars[-10:])

```

```python
def convert_to_time_of_day(readout_time_array):
    """
    Convert an array of readout timestamps to hours with fractional minutes.
    
    Parameters:
        readout_time_array (np.ndarray): Array of timestamps in seconds since epoch.
        
    Returns:
        tuple: A tuple containing:
            - list: List of times in hours as floats.
            - list: List of corresponding datetime objects.
    """
    # Convert readout_time to datetime objects
    times_of_day = [datetime.fromtimestamp(t) for t in readout_time_array]
    
    # Extract hours with fractional minutes
    readout_hours = [
        t.hour + t.minute / 60 + t.second / 3600 for t in times_of_day
    ]
    
    return readout_hours, times_of_day  # Ensure we are returning both lists
```

```python

def convert_to_time_of_day(readout_time_array,reference_t,base_time=datetime(2024, 1, 1, 0, 0, 30)):
    """
    Convert an array of readout timestamps to hours with fractional minutes.
    
    Parameters:
        readout_time_array (np.ndarray): Array of timestamps in seconds since epoch.
        
    Returns:
        tuple: A tuple containing:
            - list: List of times in hours as floats.
            - list: List of corresponding datetime objects.
    """
    # Convert readout_time to datetime objects
    times_of_day = [datetime.fromtimestamp(t) - reference_t + base_time  for t in readout_time_array]

    return times_of_day

def plot_event_data(event_day_arr):
    plt.figure(figsize=(12, 6))  # Create a larger figure
    ax1 = plt.gca()  # Get the current axes

    date_labels = []  # To store median dates for generating start and end times
    base_time = datetime.fromtimestamp(event_day_arr[0].readout_time[0])
    for event in event_day_arr:
        # Ensure event.readout_time is an array-like structure
        if hasattr(event, 'readout_time') and isinstance(event.readout_time, (np.ndarray, list)):
            # Convert readout_time to hours and datetime objects
            reference_t = datetime.fromtimestamp(event.readout_time[0])
            readout_hours = convert_to_time_of_day(event.readout_time, reference_t, base_time)

            # Calculate the median timestamp for the legend label
            median_timestamp = np.median(event.readout_time)
            date_label = datetime.fromtimestamp(median_timestamp).strftime("%b %d")
            date_labels.append(datetime.fromtimestamp(median_timestamp).strftime("%Y-%m-%d"))

            # Extract lower and upper error bars
            lower_error = event.error_bar[:, 0]
            upper_error = event.error_bar[:, 1]
            asym_error = [lower_error, upper_error]  # Asymmetric error bars

            # Plot the data with error bars
            ax1.errorbar(
                readout_hours, event.V_rms, yerr=asym_error,
                fmt='o', capsize=5, label=date_label
            )
        else:
            print("Invalid readout_time for event.")

    # Set x-axis formatting to show hours and minutes only
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

    # Generate start and end times based on the first and last dates
    start_time_i = f"{date_labels[0]} 00:00:30"
    end_time_i = f"{date_labels[0]} 23:59:30"
    start_time_f = f"{date_labels[-1]} 00:00:30"
    end_time_f = f"{date_labels[-1]} 23:59:30"
    
    # Get galaxy altitudes using the assumed function
    altitudes_i, times_i = get_galaxy_altitudes(start_time_i, end_time_i, 1./49)
    altitudes_f, times_f = get_galaxy_altitudes(start_time_f, end_time_f, 1./49)

    formatted_time_i = times_i.to_datetime()
    formatted_time_f = times_f.to_datetime() - timedelta(days=7)

    # Set axis labels, title, and grid for primary y-axis
    ax1.set_xlabel("Time of Day (Hours:Minutes)")
    ax1.set_ylabel("V_rms")
    ax1.set_title("V_rms Measurements throughout the Day")
    ax1.grid(True)

    # Initialize lists to collect legend handles and labels
    handles, labels = ax1.get_legend_handles_labels()

    # Create a secondary y-axis (right) for altitudes (initial time period)
    ax2 = ax1.twinx()
    line1, = ax2.plot(formatted_time_i, altitudes_i[:49], 'green', label='Galaxy Altitude at Start Date')
    ax2.set_ylabel('Altitude')
    ax2.tick_params(axis='y', labelcolor='green')

    # Add the handle and label for ax2 to the legend lists
    handles.append(line1)
    labels.append('Galaxy Altitude at Start Date')

    # Create another secondary y-axis (right) for altitudes (final time period)
    ax3 = ax1.twinx()
    ax3.spines['right'].set_position(('outward', 60))  # Offset the third axis
    line2, = ax3.plot(formatted_time_f, altitudes_f, 'cyan', label='Galaxy Altitude at End Date')

    # Add the handle and label for ax3 to the legend lists
    handles.append(line2)
    labels.append('Galaxy Altitude at End Date')

    # Add combined legend to ax1
    ax1.legend(handles, labels, title="Event Date (Median)", loc="upper left")
    # Create the filename with start datetime and channel number
    filename = f"beacon_analysis_at_channel_{4}_for_week_of_dec_1.png"
    
    # Save the figure
    plt.savefig(filename)
    plt.show()

# Call the plot function
plot_event_data(event_day_arr)
```

```python
# Example usage
# start_time_i = '2024-02-14 00:00:41'
# end_time_i = '2024-02-15 00:00:40'
start_time_i = '2023-11-30 00:00:34'
end_time_i = '2023-12-1 00:00:30'

start_time_f = '2023-12-07 00:00:50'
end_time_f = '2023-12-08 00:00:30'
altitudes_i, times_i = get_galaxy_altitudes(start_time_i, end_time_i,1./49)
altitudes_f, times_f = get_galaxy_altitudes(start_time_f, end_time_f,1./49)

time_gap=36*60


#function use to check the mismatch of the time which might cause problem for average the data
def check_break_time(event_day_array):
    for event_day in event_day_array:
        previous_time = event_day.readout_time[0]
        for i in range(len(event_day.readout_time)):
            time=event_day.readout_time[i]
            if time-previous_time > time_gap:
                print("gap more than 36min!")
            previous_time=time
        print("check finished")

def average_event(event_day_array):
    readout_time_bundle = []
    V_rms_bundle = []
    upper_error_bar_bundle = []
    lower_error_bar_bundle = []
    V_rms_20_bundle = []
    
    for event in event_day_array:
        readout_time_bundle.append(event.readout_time)
        V_rms_bundle.append(event.V_rms)
        upper_error_bar_bundle.append(event.error_bar[:, 0])  # Upper errors (first column)
        lower_error_bar_bundle.append(event.error_bar[:, 1])  # Lower errors (second column)
        V_rms_20_bundle.append(event.V_rms_20)
    
    # Convert bundles to numpy arrays
    readout_time_bundle = np.array(readout_time_bundle)
    V_rms_bundle = np.array(V_rms_bundle)
    upper_error_bar_bundle = np.array(upper_error_bar_bundle)
    lower_error_bar_bundle = np.array(lower_error_bar_bundle)
    V_rms_20_bundle = np.array(V_rms_20_bundle)
    
    # Calculate the mean of each parameter
    readout_time_avg = np.mean(readout_time_bundle, axis=0)
    V_rms_avg = np.mean(V_rms_bundle, axis=0)
    V_rms_20_avg = np.mean(V_rms_20_bundle, axis=0)
    
    # Calculate the combined upper and lower errors using the propagation of uncertainty formula
    upper_error_bar_avg = np.sqrt(np.sum(upper_error_bar_bundle**2, axis=0)) / upper_error_bar_bundle.shape[0]
    lower_error_bar_avg = np.sqrt(np.sum(lower_error_bar_bundle**2, axis=0)) / lower_error_bar_bundle.shape[0]
    
    # Combine upper and lower errors back into the original form
    error_bar_avg = np.column_stack((upper_error_bar_avg, lower_error_bar_avg))

    avg_event=Event_day(readout_time_avg,V_rms_avg,error_bar_avg,V_rms_20_avg)
    return avg_event


#check_break_time(event_day_arr)
#event_avg=average_event(event_day_arr)

```

```python
import matplotlib.dates as mdates
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import numpy as np

# t=event_avg.readout_time
# V_rms_20=event_avg.V_rms_20
# V_rms=event_avg.V_rms
# error_bars=event_avg.error_bar
t=event_day_arr[0].readout_time
V_rms_20=event_day_arr[0].V_rms_20
V_rms=event_day_arr[0].V_rms
error_bars=event_day_arr[0].error_bar



# Define the sine wave function
def sine_wave(t, A, T, phi, C):
    return A * np.sin(2 * np.pi * t / T + phi) + C

# Improved initial guesses and bounds
initial_guess = [5, 24*3600, 0, 7.5]  # Example initial guesses
bounds = (
    [0, 0, 0, 0],                    # Lower bounds for A, T, phi, C
    [10, 2*24*3600, 2*np.pi, 15]     # Upper bounds for A, T, phi, C
)

# Fit the curve
try:
    params, covariance = curve_fit(sine_wave, t, V_rms, p0=initial_guess, bounds=bounds, maxfev=5000)
except RuntimeError as e:
    print(f"Error occurred: {e}")
    params = [np.nan, np.nan, np.nan, np.nan]
    covariance = None

# Extract the fitted parameters
A, T, phi, C = params

# Generate fitted data
fitted_V_rms = sine_wave(t, *params)

# Convert timestamps to datetime
time_datetimes = [datetime.fromtimestamp(sec) for sec in t]

# Step 5: Plot the results using scatter plot
fig, ax1 = plt.subplots(figsize=(12, 8))

# Separate the lower and upper error bars
V_rms_lower_errors = np.array([eb[0] for eb in error_bars])
V_rms_upper_errors = np.array([eb[1] for eb in error_bars])
#V_rms_lower_errors = np.array([eb[0] - median + median_20 for eb, median, median_20 in zip(error_bars, V_rms, V_rms_20)])
#V_rms_upper_errors = np.array([eb[1] + median - median_20 for eb, median, median_20 in zip(error_bars, V_rms, V_rms_20)])

# Plot V_rms with error bars on the primary y-axis (left)
ax1.errorbar(time_datetimes, V_rms, yerr=[V_rms_lower_errors, V_rms_upper_errors], 
             fmt='o', color='b', ecolor='gray', elinewidth=1, capsize=2, label='V_rms Data')
ax1.plot(time_datetimes, fitted_V_rms, 'r-', label=f'Sine Fit: A={A:.2f}, T={T/3600:.2f} hrs, phi={phi:.2f}, C={C:.2f}')
ax1.set_xlabel('Date and Time')
ax1.set_ylabel('V_rms', color='b')
ax1.tick_params(axis='y', labelcolor='b')
#ax1.set_ylim([3,5])
# Format the x-axis as dates
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))

# Customize the tick locator for every 4 hours
ax1.xaxis.set_major_locator(mdates.HourLocator(interval=4))

# Rotate and format the x-axis labels
fig.autofmt_xdate()


y_max=90
y_min=-90




# Create a secondary y-axis (right) for altitudes
ax2 = ax1.twinx()
ax2.plot(time_datetimes, altitudes_i[:49], 'green', label='Galaxy Altitude at 2024-03-30')
ax2.set_ylabel('Altitude', color='green')
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_ylim([y_min, y_max]) 


# # Create a secondary y-axis (right) for altitudes
# ax3 = ax1.twinx()
# ax3.plot(time_datetimes, altitudes_f, 'cyan', label='Galaxy Altitude at 2024-05-14')
# ax3.set_ylabel('Altitude', color='green')
# ax3.tick_params(axis='y', labelcolSor='green')
# ax3.set_ylim([y_min, y_max]) 


# Combine legends from both axes
fig.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

plt.title('Sine Wave Fit to V_rms Data and Altitudes for channel 10')

# Create the filename with start datetime and channel number
filename = f"STD_plot_{time_datetimes[0]}_channel_{10}_random_median.png"

# Save the figure
plt.savefig(filename)
plt.show()

```
